## feature selection private

```python
x = df.drop(['UA_Name', 'UA_Country', 'UA_Continent', 'Quality of Life Index'], axis = 1)
y = df['Quality of Life Index']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
feature1 = feature(df.drop(categorail, axis = 1), 'Quality of Life Index')
m1 = feature1.mutual_info_selector(x_train, y_train, 17)
f1 = feature1.f_test_selector(x_train, y_train, 17)
sh1 = feature1.shapley_selection(x_train, y_train, x_test, y_test, LinearRegression(), 17)
pm1 = feature1.permutation_check(x_train, y_train, x_test, y_test, LinearRegression(),  17)
```

```python
df1 = pd.merge(m1, f1, left_index = True, right_index = True)
df1 = pd.merge(df1, pm1, left_index = True, right_index = True)
df1 = pd.merge(df1, sh1, left_index = True, right_index = True)

df1
```

```python
df1 = feature.overall_selector(df1, 0.05, 0.05 / 17, 0.05, 1)
def selector_c(x1, x2, x3, x4):
    if sum([x1, x2, x3, x4]) >= 3:
        return 1
    else: return 0
df1['gotIt'] = df1.apply(lambda x: selector_c(x.imp_mean_thr, x.f_p_values_thr, x.mutual_score_thr, x.shapley_mean_thr), axis = 1)
good_values = df1[df1.gotIt == 1].index
```

```python
x = df.drop(["Quality of Life Index"], axis = 1)
x = x[good_values]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
LinReg = LinearRegression()
LinReg.fit(x_train, y_train)
```

```python
selected_score = LinReg.score(x_train, y_train), LinReg.score(x_test, y_test)
selected_score
```

```python
selected_r2 = r2_score(LinReg.predict(x_train), y_train), r2_score(LinReg.predict(x_test), y_test)
selected_r2
```

```python
selected_mse = mean_squared_error(LinReg.predict(x_train), y_train), mean_squared_error(LinReg.predict(x_test), y_test)
selected_mse
```

```python
selected_score[1] > base_score[1], selected_r2[1] > base_r2[1], selected_mse[1] < base_mse[1], 
```

Удаление оправдано. Объединяем модели
