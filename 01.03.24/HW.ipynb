```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from feature_extraction import feature
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/katarina74/ml_lessons/main/lesson_2/data/techparams_train.csv")
df = df.drop(['index'], axis = 1)
x = df.drop(["target"], axis = 1)
y = df[["target"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
feature1 = feature(df, 'target')
m1 = feature1.mutual_info_selector(x_train, y_train, 20)
f1 = feature1.f_test_selector(x_train, y_train, 20)
sh1 = feature1.shapley_selection(x_train, y_train, x_test, y_test, LinearRegression(), 20)
pm1 = feature1.permutation_check(x_train, y_train, x_test, y_test, LinearRegression(),  20)
```

```python
df1 = pd.merge(m1, f1, left_index = True, right_index = True)
df1 = pd.merge(df1, pm1, left_index = True, right_index = True)
df1 = pd.merge(df1, sh1, left_index = True, right_index = True)

df1
```

```python
feature.overall_selector(df1, 0.0005, 0.05 / 32, 0.05, 1)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2)
```

```python
data1 = poly.fit_transform(x)

features = pd.DataFrame(data1, columns = poly.get_feature_names_out())
features['target'] = df.target
```

```python
x = features.drop(["target"], axis = 1)
y = features[["target"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
feature1 = feature(features, 'target')
m1 = feature1.mutual_info_selector(x_train, y_train, 20)
f1 = feature1.f_test_selector(x_train, y_train, 20)
sh1 = feature1.shapley_selection(x_train, y_train, x_test, y_test, LinearRegression(), 20)
pm1 = feature1.permutation_check(x_train, y_train, x_test, y_test, LinearRegression(),  20)
```

```python
df1 = pd.merge(m1, f1, left_index = True, right_index = True)
df1 = pd.merge(df1, pm1, left_index = True, right_index = True)
df1 = pd.merge(df1, sh1, left_index = True, right_index = True)

df1
```

```python
df1 = feature.overall_selector(df1, 0.0005, 0.05 / 32, 0.05, 1)
```

```python
def selector_c(x1, x2, x3, x4):
    if sum([x1, x2, x3, x4]) >= 3:
        return 1
    else: return 0
```

```python
df1['gotIt'] = df1.apply(lambda x: selector_c(x.imp_mean_thr, x.f_p_values_thr, x.mutual_score_thr, x.shapley_mean_thr), axis = 1)
```

```python
good_values = df1[df1.gotIt == 1].index
```

```python
x = features.drop(["target"], axis = 1)
x = x[good_values]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
LinReg = LinearRegression()
LinReg.fit(x_train, y_train)
```

```python
LinReg.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2_score(LinReg.predict(x_train), y_train)
```

```python
r2_score(LinReg.predict(x_test), y_test)
```
